```python
corpus=["Paragraphs are the building blocks of papers.", "Many students define paragraphs in terms of length: a paragraph is a group of at least five sentences, a paragraph is half a page long, etc.",
        "In reality, though, the unity and coherence of ideas among sentences is what constitutes a paragraph.", "A paragraph is defined as “a group of sentences or a single sentence that forms a unit” (Lunsford and Connors 116).",
        "Length and appearance do not determine whether a section in a paper is a paragraph.", "For instance, in some styles of writing, particularly journalistic styles, a paragraph can be just one sentence long.", "Ultimately, a paragraph is a sentence or group of sentences that support one main idea.",
        "In this handout, we will refer to this as the “controlling idea",
         "because it controls what happens in the rest of the paragraph."]

```

```python
print(corpus)
```

```python
len(corpus)
```

```python
corpus[8]
```

```python
token=[]
for paragraph in corpus:
  for word in paragraph.split(" "):
    token.append(word)
```

```python
print(token)
```

```python
len(token)
```

```python
vocab = {"<pad>": 0}
idx_counter = 1
for word in token:
  if word not in vocab:
    vocab[word] = idx_counter
    idx_counter += 1
```

```python
len(vocab)
```

```python
sequence=[vocab[word] for word in token]
```

```python
sequence
```

```python
vocab
```

```python
import torch
import torch.nn as nn
```

```python
embedding=nn.Embedding(
    num_embeddings=len(vocab),
    embedding_dim=5,
    padding_idx=0
)
```

```python
seq_=torch.tensor(sequence,dtype=torch.int64)

emb_seq=embedding(seq_)
```

```python
emb_seq.shape
```

```python
model=nn.RNN(input_size=5,
             hidden_size=20)
```

```python
output, hidden = model(emb_seq)
print(f"Output shape: {output.shape}")
print(f"Hidden state shape: {hidden.shape}")
```
