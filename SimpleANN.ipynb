```python
import torch
import torch.nn as nn
```

```python
class SimpleANN(nn.Module):
  def __init__(self):
    super(SimpleANN,self).__init__()
    self.layer1=nn.Linear(3,5)
    self.relu=nn.ReLU()
    self.layer2=nn.Linear(5,2)
    self.sigmoid=nn.Sigmoid()


  def forward(self,x):
      output_from_layer_one=self.layer1(x)
      activation_from_layer_one=self.relu(output_from_layer_one)
      output_from_layer_two=self.layer2(activation_from_layer_one)
      activation_from_layer_two=self.sigmoid(output_from_layer_two)
      return activation_from_layer_two


#main class
x=torch.tensor([1,2,3],dtype=torch.float32)

model=SimpleANN()
output=model(x)

print(output)






```

```python
# Create the ANN from the scratch
class Activation:
  def __init__(self):
    pass

  def sigmoid(self,value):
    return 1/(1+torch.exp(-value))

  def Relu(self,value):
    # Use torch.relu for element-wise operation on tensors
    return torch.relu(value)

activa=Activation()

class ANNFROMSCRATCH:
  def __init__(self):
    pass


  def forwardlayer1(self,wight1,biase1,inputfeature,hiddenneuron):
    self.wight1=wight1
    self.biase1=biase1
    self.inputfeature=inputfeature
    self.hiddenneuron=hiddenneuron
    self.x=inputfeature @ wight1.T +biase1
    self.outputlayer1=activa.Relu(self.x)
    return self.outputlayer1


  def forwardlayer2(self,wight2,biase2,inputfeature2,hiddenneuron2):
    self.wight1=wight2
    self.biase1=biase2
    self.inputfeature=inputfeature2
    self.hiddenneuron=hiddenneuron2
    self.x=inputfeature2 @ wight2.T +biase2
    self.outputlayer2=activa.Relu(self.x)
    return self.outputlayer2


  def predictlayer(self,sigmoid1):
   self.sig=activa.sigmoid(sigmoid1)
   return self.sig










#main

input_fea=torch.tensor([1,2,3],dtype=torch.float32)
hidden_neu=5
wight1=torch.ones(hidden_neu,len(input_fea),dtype=torch.float32)
biase1=torch.ones(5,dtype=torch.float32)
object1=ANNFROMSCRATCH()

output_from_layer1=object1.forwardlayer1(wight1,biase1,input_fea,hidden_neu)
hidden_neu2=2
wight2=torch.ones(hidden_neu2,hidden_neu,dtype=torch.float32)
biase2=torch.ones(hidden_neu2,dtype=torch.float32)
output_from_layer2=object1.forwardlayer2(wight2,biase2,output_from_layer1,hidden_neu2)
print(object1.predictlayer(output_from_layer2))
```
