```python
import torch
import torch.nn as nn
```

```python
class SimpleANN(nn.Module):
  def __init__(self):
    super(SimpleANN,self).__init__()
    self.layer1=nn.Linear(3,5)
    self.relu=nn.ReLU()
    self.layer2=nn.Linear(5,2)
    self.sigmoid=nn.Sigmoid()


  def forward(self,x):
      output_from_layer_one=self.layer1(x)
      activation_from_layer_one=self.relu(output_from_layer_one)
      output_from_layer_two=self.layer2(activation_from_layer_one)
      activation_from_layer_two=self.sigmoid(output_from_layer_two)
      return activation_from_layer_two


#main class
x=torch.tensor([1,2,3],dtype=torch.float32)

model=SimpleANN()
output=model(x)

print(output)






```
